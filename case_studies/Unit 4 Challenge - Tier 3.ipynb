**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# Check data type of properties & set as DataFrame if it's not already
type(properties) # DataFrame

# Drop the 3 blank columns and the columns that aren't boroughs of London
# These are all the columns from Barking & Dagenham to Westminster 
properties = properties.loc[:,:'Westminster']
properties = properties.drop('City of London',axis=1)
properties.head()
```

```python
# The top row look like an ID for the borough, but I'm going to use the borough name to my identifier.
# I don't want to lose track of the IDs, but I want them gone for now, so I'm going to subset them
# and save them for later as borough_IDs. Then I'll drop that row from properties. 
borough_IDs = properties.iloc[0]  # Extract first row
borough_IDs = pd.DataFrame(borough_IDs)  # Convert to DataFrame

# Rename column
borough_IDs.columns = ['borough_ID']

# Reset index to make 'borough' a column
borough_IDs = borough_IDs.reset_index().rename(columns={'index': 'borough'})

# Drop the first row of data if needed
borough_IDs = borough_IDs.iloc[1:]

borough_IDs.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Drop first row 
properties = properties.iloc[1:]
#  Rename the date column so it becomes the column names when we transpose
properties.rename(columns={"Unnamed: 0": "date"},inplace=True)
properties.head()
```

```python
# Transpose the properties so the borough is one column and the dates are the column names
properties_transposed = properties.set_index("date").transpose().reset_index().rename(columns={"index": "borough"})
properties_transposed.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Now melt the dataset into long form
properties_long = properties_transposed.melt(id_vars=["borough"], var_name="date", value_name="average_price")
properties_long.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
properties_long["date"] = pd.to_datetime(properties_long["date"])
properties_long["average_price"] = properties_long["average_price"].astype(float)
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit.

```python
# We're asked to evaluate prices over the last two decades, but we have nearly 3 decades of data
# Subset properties_long to focus on prices from 2004-12-01 to 2024-12-01 (our max date)
# All boroughs have the same max date (2024-12-01)
from dateutil.relativedelta import relativedelta

min_date = max(properties_long['date']) - relativedelta(years = 20)
properties_two_decades = properties_long[properties_long['date'] > min_date].sort_values('date')
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

```python
# Add a column that gives the average price by year for each borough
properties_two_decades["yearly_avg_price"] = properties_two_decades.groupby(["borough", "year"])["average_price"].transform(lambda x: x.mean())
properties_two_decades.head(5)
```

```python
# Compare prices between 2004 and 2024. Filter out columns we don't need to focus on.
recent_prices = properties_two_decades[['borough','yearly_avg_price']][properties_two_decades["year"] == max(properties_two_decades["year"])].drop_duplicates()
prices_20_yrs_ago = properties_two_decades[['borough','yearly_avg_price']][properties_two_decades["year"] == min(properties_two_decades["year"])].drop_duplicates()

price_comparison = recent_prices.merge(prices_20_yrs_ago,on='borough',how='left',suffixes=('_2024','_2004'))
price_comparison['percent_change'] = ((price_comparison['yearly_avg_price_2024'] - price_comparison['yearly_avg_price_2004'])/price_comparison['yearly_avg_price_2004'])*100
print(price_comparison[['borough','percent_change']].sort_values('percent_change',ascending=False).reset_index(drop=True))
```

```python
%matplotlib inline 
price_comparison = price_comparison.sort_values('percent_change')

# Create the lollipop chart
fig, ax = plt.subplots()

# Plot the dots
ax.scatter(price_comparison['percent_change'], price_comparison['borough'])

# Add labels and title
ax.set_xlabel('Price Change (%)')
ax.set_ylabel('Borough')
ax.tick_params(axis='y', labelsize=8)
ax.set_title('Change in Average Housing Price, 2004-2024')
ax.grid(True, linewidth=0.5,alpha=0.3)
ax.axvline(x=100, color='red', linestyle='--', linewidth=1)

plt.show()
```
